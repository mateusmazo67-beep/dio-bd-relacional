```
import google.generativeai as genai
from google.colab import userdata

# Obtém a chave da API do Gemini dos segredos do Colab
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Inicializa o modelo Gemini com a versão 'latest'
gemini_model = genai.GenerativeModel('gemini-pro-latest')
```

```
# Removendo imports e inicialização do OpenAI
# O 'gemini_model' é inicializado globalmente em uma célula anterior.

def generate_ai_news(user):
  # Criando o prompt para o Gemini
  # O prompt foi ajustado para ser mais direto para o Gemini e evitar a sintaxe de 'role' específica do OpenAI ChatCompletion.
  prompt_content = f"Você é um personagem de Naruto. Crie uma mensagem para {users['name']} falando sobre o poder da amizade, se identificando como um personagem de Naruto. A mensagem deve ter no máximo 100 caracteres."

  response = gemini_model.generate_content(prompt_content)

  # A resposta do Gemini é acessada diretamente via .text
  return response.text.strip('"')

for users in users:
  news = generate_ai_news(user)
  print(news)
  user['news'].append({
      "icon":
      "description": news
  })
```
